## Probability: The Foundation of Language Models
### What Is a Language Model?

**Definition:** A model that assigns probabilities to sequences of words (or tokens).

**Example:** Given the context "The cat sat on the", what is the probability of the next word?

```
P("mat"   | "The cat sat on the") = 0.35
P("couch"  | "The cat sat on the") = 0.25
P("floor" | "The cat sat on the") = 0.20
P("roof"  | "The cat sat on the") = 0.10
P(other)  = 0.10
```

Key property: The probabilities over all possible next tokens must sum to 1.0, forming a valid probability distribution.

A language model estimates the conditional probability:

$$
P(x_t \mid x_{<t})
$$

**Technical description.**  
The probability of the token $x_t$ at time step $t$, conditioned on the sequence of all tokens generated prior to $t$.

**Non-technical description.**  
The model reads everything that has been generated so far and assigns probabilities to each possible next token.

````python
# Context: "The cat sat on the"
possible_tokens = ["mat", "couch", "floor", "roof", "wall", "car"]

# Probabilities assigned by the model (hypothetical)
probabilities = np.array([0.35, 0.25, 0.20, 0.10, 0.07, 0.03])

# Check if sum 1.00
print(f"Probabilities sum: {probabilities.sum():.2f}")
print(f"Is a valid distribution? {np.isclose(probabilities.sum(), 1.0)}")
````

````python
plt.figure(figsize=(10, 5))
plt.bar(possible_tokens, probabilities, color='tab:blue')
plt.xlabel('Next Token', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.title('Probability Distribution', fontsize=12, fontweight='bold')
plt.ylim(0, 0.4)
````

### Interpretation:
- "mat" has a 35% probability (most likely)
- "car" has only a 3% probability (least likely)
- The model is confident that the next token is "mat"


### Important Question

**How does the neural model generate these probabilities?**

**Answer:**  
The neural model produces **scores** (arbitrary real-valued numbers), not probabilities directly.

**Problem:**  
Scores can be:
- Negative  
- Greater than 1  
- Not normalized (do not sum to 1.0)

**Solution:**  
The **Softmax** function

$$
\mathrm{Softmax}(z_i) = \frac{e^{z_i}}{\sum_{j \in \mathcal{V}} e^{z_j}}
$$

### How to read
The softmax of the score $z_i$ is equal to the exponential of $z_i$ divided by the sum of the exponentials of all scores $z_j$ in the vocabulary.

### Technical description
The Softmax function maps a vector of unnormalized real-valued scores (logits) $z$ into a categorical probability distribution over the vocabulary $\mathcal{V}$.  
Each output represents the normalized probability of selecting token $i$, ensuring all probabilities are positive and sum to one.

### Where:
- $x_i$ = score of token $i$  
- $e$ = Euler's number (~2.718)  
- Division by the sum ensures that all probabilities add up to 1.0

### Why exponential ($e^x$)?
1. **Turns negatives into positives:** $e^{-5} = 0.0067$ (small but positive)  
2. **Amplifies differences:** higher scores become more dominant  
3. **Mathematically convenient:** derivative is simple (important for backpropagation)

### Non-technical description
The model first assigns a score to each possible next token.  
Softmax turns these scores into probabilities by making them positive, comparing them to each other, and normalizing them so the total equals 1.

````python
def softmax(x):
    x_stable = x - np.max(x)
    exp_x = np.exp(x_stable)

    return exp_x / np.sum(exp_x)


# Try with arbitrary numbers
scores = np.array([2.5, 1.0, 0.5, -1.0, -0.5, 0.0])
print("Model Scores (Arbitrary Numbers) \n")
print(scores, "\n")

print(f"   Sum: {scores.sum():.2f} is not 1.00")
print(f"   Have negative numbers: {(scores < 0).any()}")

# Apply Softmax
probs = softmax(scores)

print("\nAfter Softmax (Probabilities)\n")
print(probs, "\n")
print(f"   Sum {probs.sum():.4f} (Exactly 1)")
print(f"   All positives: {(probs > 0).all()}")
print(f"   Highest probability: {probs.max():.4f} (Index: {probs.argmax()})")
````

### Transformation View

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

tokens = ['token_0', 'token_1', 'token_2', 'token_3', 'token_4', 'token_5']

# Before Softmax (Scores)

ax1.bar(tokens, scores, color='tab:orange', alpha=1)
ax1.set_ylabel('Score (Arbitrary value)', fontsize=11)
ax1.set_title('Before: Model Scores', fontsize=13, fontweight='bold')
ax1.axhline(y=0, color='black', linestyle='--', linewidth=0.8, alpha=0.3)
ax1.grid(axis='y', alpha=0.3)

# After Softmax (Probabilities)
ax2.bar(tokens, probs, color='tab:pink', alpha=1)
ax2.set_ylabel('Probabilities', fontsize=11)
ax2.set_title('After: Probabilities (Softmax)', fontsize=13, fontweight='bold')
ax2.set_ylim(0, 1)
ax2.axhline(y=1.0, color='red', linestyle='--', linewidth=0.8, alpha=0.3, label='Sum = 1.0')
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n Note:\n")
print("  1. Negative scores became small positive probabilities")
print("  2. The highest score (2.5) became the highest probability (~0.73)")
print("  3. Differences were amplified (winner-take-most)")
print("  4. Everything now sums to 1.0 (valid probability distribution)")
````

### Temperature Effect

The **temperature** controls how "confident" the model is:

$$
\text{softmax}(x_i, T) = \frac{e^{x_i/T}}{\sum_{j} e^{x_j/T}}
$$

### How to read
The softmax of score $x_i$ with temperature $T$ is equal to the exponential of $x_i$ divided by $T$, divided by the sum of the exponentials of all scores $x_j$ divided by $T$.

### Technical description
Temperature $T$ scales the logits before applying softmax.  
- $T = 1.0$: normal behavior  
- $T \to 0$: the distribution becomes very sharp (almost deterministic)  
- $T \to \infty$: the distribution becomes flat (almost uniform)  

Scaling by $1/T$ allows controlling the entropy of the output distribution: smaller $T$ reduces entropy, making the model more confident; larger $T$ increases entropy, making it more uncertain.

### Non-technical description
Temperature adjusts how sure the model is about the next token:  
- Low $T$  model is very confident, picks the highest-scoring token most of the time  
- High $T$  model is uncertain, spreads probabilities more evenly among possible tokens  
- Normal $T = 1$  standard behavior, balance between confident and diverse predictions

````python
def softmax_with_temperature(x, temperature=1.0):
    x_temp = x / temperature
    x_stable = x_temp - np.max(x_temp)
    exp_x = np.exp(x_stable)
    
    return exp_x / np.sum(exp_x)


temperatures = [0.5, 1.0, 2.0]
scores_test = np.array([3.0, 2.0, 1.0, 0.5, 0.2])

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, temp in zip(axes, temperatures):
    probs_temp = softmax_with_temperature(scores_test, temperature=temp)
    
    ax.bar(range(len(probs_temp)), probs_temp, color='tab:blue')
    ax.set_title(f'Temperature = {temp}', fontsize=12, fontweight='bold')
    ax.set_ylabel('Probabilities', fontsize=10)
    ax.set_xlabel('Token', fontsize=10)
    ax.set_ylim(0, 1)
    ax.grid(axis='y', alpha=0.3)
    
    # Show values
    for i, p in enumerate(probs_temp):
        ax.text(i, p + 0.02, f'{p:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show

print("\nTemperature Interpretation:")
print("   T=0.5 (low):     Highly focused on the most probable token (deterministic)")
print("   T=1.0 (normal):  Balance between confidence and diversity")
print("   T=2.0 (high):    More spread out, more 'creative' (random)")

print("\nPractical usage:")
print("  - Code generation: T=0.2-0.5 (needs to be precise)")
print("  - Creative writing: T=0.7-1.0 (needs to be interesting)")
print("  - Brainstorming: T=1.2-1.5 (needs to be diverse)")
````

## Cross-Entropy Loss: Measuring the Error

### The Problem
We have:
- **Model prediction:** Probability distribution over tokens  
- **Reality (label):** Which token is actually correct  

**How do we measure how wrong the model is?**

### The Formula

$$
\text{CrossEntropy} = -\sum_{i} y_{\text{true},i} \cdot \log(y_{\text{pred},i})
$$

For classification (where only one token is correct), this simplifies to:

$$
\text{Loss} = -\log(P(\text{correct token}))
$$

### How to read
The cross-entropy loss is the negative log of the predicted probability assigned to the true token (or the sum over all tokens weighted by the true distribution).

### Technical description
Cross-entropy quantifies the difference between the predicted probability distribution $y_{\text{pred}}$ and the true distribution $y_{\text{true}}$.  
- If the model assigns low probability to the correct token, the loss is high  
- If the model assigns high probability to the correct token, the loss is low  
- Widely used for training classification models via gradient descent

### Non-technical description
Think of it as a **penalty for being wrong**:  
- Predicting a token confidently but incorrectly  big penalty  
- Predicting the correct token confidently  small penalty  
- The logarithm ensures the penalty scales nicely and is mathematically convenient

### Why the logarithm?
1. **Penalizes confident mistakes:** Predicting 1% when it should be 100%  high loss  
2. **Rewards confident correct predictions:** Predicting 99% for the right token  low loss  
3. **Mathematical convenience:** $\log(e^x) = x$ (cancellation in derivatives)


````python
def cross_entropy_loss(predictions, target_index):
    
    # Adding epsilon to avoid log(0)
    epsilon = 1e-10

    correct_prob = predictions[target_index]

    return -np.log(correct_prob + epsilon)


### Examples

target = 0  # Correct token is 0

print("Scenario 1: Very good model")

probs_good = np.array([0.95, 0.03, 0.01, 0.01])  # Very confident in token 0
loss_good = cross_entropy_loss(probs_good, target)

print(f"  Predicted probability for correct token: {probs_good[target]:.2%}")
print(f"  Loss: {loss_good:.4f} (EXCELLENT!)")

print("\nScenario 2: Reasonable model")

probs_ok = np.array([0.50, 0.30, 0.15, 0.05])  # Moderately confident
loss_ok = cross_entropy_loss(probs_ok, target)

print(f"  Predicted probability for correct token: {probs_ok[target]:.2%}")
print(f"  Loss: {loss_ok:.4f} (OK)")

print("\nScenario 3: Bad model")

probs_bad = np.array([0.10, 0.60, 0.20, 0.10])  # Predicted wrong token as most probable
loss_bad = cross_entropy_loss(probs_bad, target)

print(f"  Predicted probability for correct token: {probs_bad[target]:.2%}")
print(f"  Loss: {loss_bad:.4f} (BAD!)")

print("\nScenario 4: Very bad model")

probs_very_bad = np.array([0.01, 0.33, 0.33, 0.33])  # Almost ignoring the correct token
loss_very_bad = cross_entropy_loss(probs_very_bad, target)

print(f"  Predicted probability for correct token: {probs_very_bad[target]:.2%}")
print(f"  Loss: {loss_very_bad:.4f} (VERY BAD!)")
````

````python
### Generating the Probability vs. Loss Curve

# Test probabilities
probabilities_test = np.linspace(0.01, 1.0, 100)
losses = -np.log(probabilities_test)

plt.figure(figsize=(12, 6))

# Plot the curve
plt.plot(probabilities_test, losses, linewidth=3, color='tab:blue', label='Loss = -log(p)')

# Tab10 colors for important points
tab_colors = ['tab:green', 'tab:orange', 'tab:red', 'tab:purple']
important_points = [
    (0.95, -np.log(0.95), 'Excellent\n(p=95%)', tab_colors[0]),
    (0.50, -np.log(0.50), 'Reasonable\n(p=50%)', tab_colors[1]),
    (0.10, -np.log(0.10), 'Bad\n(p=10%)', tab_colors[2]),
    (0.01, -np.log(0.01), 'Very Bad\n(p=1%)', tab_colors[3]),
]

for p, loss, label, color in important_points:
    plt.scatter([p], [loss], s=150, c=color, zorder=5, edgecolors='black', linewidth=2)
    plt.annotate(label, xy=(p, loss), xytext=(p + 0.05, loss + 0.3),
                 fontsize=10, fontweight='bold', color=color,
                 arrowprops=dict(arrowstyle='->', color=color, lw=1.5))

plt.xlabel('Predicted Probability for Correct Token', fontsize=12)
plt.ylabel('Cross-Entropy Loss', fontsize=12)
plt.title('How Loss Penalizes Wrong Predictions', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.ylim(0, 5)

plt.tight_layout()
plt.show()

print("\nCurve Interpretation:")
print("   The curve is exponential (log is the inverse of exp)")
print("   When p  1 (100% correct): Loss  0 (perfect!)")
print("   When p  0 (0% correct): Loss   (very bad!)")
print("   Loss penalizes a lot when the model is confident but wrong")

print("\nTraining Objective:")
print("  Minimize the loss = increase the probability of the correct token")
````

## Putting It All Together: Complete Forward Pass

We will simulate how a neural model processes input until the loss is computed.

**Pipeline:**

Input (tokens)  Neural Model  Scores (logits)  Softmax  Probabilities  Loss

### How to read
- Tokens are converted into numerical representations and fed into the model.  
- The model outputs **scores** (logits) for each possible next token.  
- **Softmax** converts logits into a valid probability distribution.  
- **Cross-entropy loss** measures how well the predicted probabilities match the correct token.

### Technical description
1. **Input tokens:** Numerical encoding of the text.  
2. **Neural model:** Computes a vector of unnormalized scores for each token in the vocabulary.  
3. **Softmax:** Converts logits into probabilities:  
   $$
   P(x_i) = \frac{e^{z_i}}{\sum_{j \in \mathcal{V}} e^{z_j}}
   $$  
4. **Loss:** Cross-entropy quantifies the difference between predicted distribution and true token:  
   $$
   \text{Loss} = -\log(P(\text{correct token}) + \epsilon)
   $$  

### Non-technical description
- The model "looks" at all previous tokens.  
- It decides **how likely each possible next token is**.  
- Softmax turns raw scores into percentages.  
- Loss tells us **how wrong or right the model was**, guiding training to improve predictions.


````python
def simulated_neural_model(input_text, vocab_size = 10):
    # Simulate scores (normally obtained from network layers)
    return np.random.randn(vocab_size)

def softmax(scores):
    e_scores = np.exp(scores - np.max(scores))
    return e_scores / e_scores.sum()

def cross_entropy_loss(probs, target_idx, epsilon = 1e-8):
    return -np.log(probs[target_idx] + epsilon)

# Complete simulation
print("SIMULATION OF COMPLETE FORWARD PASS")
print("=" * 60)

# 1. Input
input_text = "The cat sat on"  # In practice, tokenized
vocab = ['mat', 'couch', 'floor', 'roof', 'wall', 'car', 'bench', 'garden', 'ceiling', 'tile']
correct_token = 'mat'  # True label
target_idx = vocab.index(correct_token)

print(f"\nINPUT: '{input_text}'")
print(f"Vocab: {vocab}")
print(f"   Correct token: '{correct_token}' (index {target_idx})")

# 2. Neural model produces scores
scores = simulated_neural_model(input_text, vocab_size=len(vocab))
print(f"\n NEURAL MODEL  Scores:")
for i, (token, score) in enumerate(zip(vocab, scores)):
    marker = "*" if i == target_idx else " "
    print(f"   {marker} {token:10s}: {score:+.3f}")


# 3. Softmax converts scores to probabilities
probs = softmax(scores)
print(f"\nSOFTMAX  Probabilities:")
for i, (token, prob) in enumerate(zip(vocab, probs)):
    marker = "*" if i == target_idx else " "
    bar = "" * int(prob * 50)  # Visual bar
    print(f"   {marker} {token:10s}: {prob:.4f} {bar}")
print(f"   Sum: {probs.sum():.10f} ")


# 4. Calculate loss
loss = cross_entropy_loss(probs, target_idx)
print(f"\nCROSS-ENTROPY LOSS:")
print(f"   Probability for '{correct_token}': {probs[target_idx]:.4f}")
print(f"   Loss: {loss:.4f}")

if loss < 0.5:
    print(f"   Evaluation: EXCELLENT. Model very confident.")
elif loss < 1.5:
    print(f"   Evaluation: GOOD. Model reasonably confident.")
elif loss < 3.0:
    print(f"   Evaluation: FAIR. Model uncertain.")
else:
    print(f"   Evaluation: BAD. Model performing poorly.")


print("\n" + "=" * 60)
print("During training, model weights are adjusted")
print("to MINIMIZE this loss (increase probability of the correct token)")
````

## Backpropagation: How the Model Learns

### Intuitive Concept

**Problem:** The model has a high loss. How do we adjust its weights to reduce the loss?

**Solution:** **Gradient Descent**

**Analogy:** Imagine you are on a mountain blindfolded and want to go down (minimize loss):
1. Feel the slope of the terrain (gradient)
2. Take a step in the downhill direction
3. Repeat until reaching the bottom (minimum loss)

### The Math (Simplified)

For each weight $w$ of the model:

$$
w_{\text{new}} = w_{\text{old}} - \alpha \cdot \frac{\partial L}{\partial w}
$$

Where:
- $\alpha$ = **learning rate** (step size)  
- $\frac{\partial L}{\partial w}$ = **gradient** (direction to reduce loss)

### Non-technical description
- The gradient tells the model **which way to adjust each weight** to reduce error.  
- Learning rate controls **how big each adjustment step is**.  
- Repeating this process iteratively allows the model to **improve its predictions** over time.

````python
# Simple loss function (parabola)
def loss_function(w):
    """Loss = (w - 3)^2 (minimum at w=3)"""
    return (w - 3.0) ** 2

# Gradient of the function
def gradient(w):
    """Derivative of (w-3)^2 = 2*(w-3)"""
    return 2 * (w - 3.0)

# Gradient descent optimizer
def gradient_descent(initial_w, learning_rate, n_steps):
    """
    Optimization using gradient descent.
    
    Parameters:
        initial_w (float): Initial weight (random)
        learning_rate (float): Step size
            - Too high: can diverge (big jumps)
            - Too low: converges slowly
            - Ideal: problem-dependent (typically 0.001 - 0.1)
        n_steps (int): Number of optimization steps
        
    Returns:
        np.ndarray: History of weights and corresponding losses
    """
    w = initial_w
    history = [(w, loss_function(w))]
    
    for step in range(n_steps):
        grad = gradient(w)
        w = w - learning_rate * grad
        history.append((w, loss_function(w)))
        
    return np.array(history)

# Test different learning rates
learning_rates = [0.01, 0.1, 0.5]
initial_w = 0.0
n_steps = 20

tab_colors = ['tab:blue', 'tab:orange', 'tab:green']

plt.figure(figsize=(15, 5))

for idx, (lr, color) in enumerate(zip(learning_rates, tab_colors), 1):
    plt.subplot(1, 3, idx)
    
    # Plot loss function
    w_range = np.linspace(-1, 7, 200)
    losses = [loss_function(w) for w in w_range]
    plt.plot(w_range, losses, color=color, alpha=0.3, linewidth=2, label='Loss function')
    
    # Run gradient descent
    history = gradient_descent(initial_w, lr, n_steps)
    
    # Plot trajectory
    plt.plot(history[:, 0], history[:, 1], 'o-', color=color, markersize=8, 
             linewidth=2, alpha=0.7, label='Optimization')
    
    # Mark start and end
    plt.scatter([history[0, 0]], [history[0, 1]], s=200, c='tab:green', 
                marker='*', zorder=5, edgecolors='black', linewidth=2, label='Start')
    plt.scatter([history[-1, 0]], [history[-1, 1]], s=200, c='tab:red', 
                marker='X', zorder=5, edgecolors='black', linewidth=2, label='End')
    
    plt.xlabel('Weight (w)', fontsize=11)
    plt.ylabel('Loss', fontsize=11)
    plt.title(f'Learning Rate = {lr}', fontsize=12, fontweight='bold')
    plt.legend(fontsize=9)
    plt.grid(True, alpha=0.3)
    
    # Final info
    final_w = history[-1, 0]
    final_loss = history[-1, 1]
    plt.text(0.5, 0.95, f'Final: w={final_w:.3f}, loss={final_loss:.3f}',
             transform=plt.gca().transAxes, fontsize=9,
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

print("\nAnalysis of Results:")
print("\n  LR = 0.01 (Low):")
print("    - Stable, no oscillations")
print("    - Converges very slowly (still far from optimum)")
print("\n  LR = 0.1 (Medium):")
print("    - Good balance: fast and stable")
print("    - Converges near optimum (w3, loss0)")
print("\n  LR = 0.5 (High):")
print("    - Converges very quickly")
print("    - Can oscillate (big jumps)")
print("\nChoosing the right learning rate is critical for training!")
````
